```python
# 데이터 불러오기
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df = pd.read_csv("https://raw.githubusercontent.com/rasbt/python-machine-learning-book/master/code/datasets/housing/housing.data", delimiter=r"\s+", 
                names = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"])
# df.head()
# df.shape
df[pd.isnull(df).any(axis=1)]
# pd.isnull(df).any(axis=1)
X = df[["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"]]
y = df[["MEDV"]]
```

```python
# Train Test Set 나누기
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size = 0.20, random_state=5)
```

```python
# Gradient Descent 함수 생성
def gradientDescent(x, y, theta, alpha, n, numIterations):
    tmp = x.copy()
    tmp = (tmp - tmp.mean()) / (tmp.max() - tmp.min())
    tmp.insert(0,"bias", 1)
    x = np.array(tmp)
    y = np.array(y).ravel()
    # bias 추가
    xTrans = x.transpose()
    for i in range(0, numIterations):
        hypothesis = np.dot(x, theta)
        loss = hypothesis - y
        # avg cost per example (the 2 in 2*n doesn't really matter here.
        # But to be consistent with the gradient, I include it)
        cost = np.sum(loss ** 2) / (2 * n)
        if(i%1000==0):
            print("Iteration %d | Cost: %f" % (i, cost))
        # avg gradient per example
        gradient = np.dot(xTrans, loss) / n
        # update
        theta = theta - alpha * gradient
    return theta
```

```python
# 예측 함수 생성
def predict(x, theta):
    return np.dot(x, theta)
```

```python
# 학습
x = X_train
y = y_train
n, m = np.shape(x)
numIterations= 300000
alpha = 0.01
theta = np.ones(m+1)
theta = gradientDescent(x, y, theta, alpha, n, numIterations)
print(theta)
```

```python
# 예측
predict([1,0.03237, 0.0, 2.18, 0, 0.458, 6.998, 45.8, 6.0622, 3, 222.0, 18.7, 394.63, 2.94], theta)
```
